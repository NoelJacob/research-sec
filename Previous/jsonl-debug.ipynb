```python
import pandas as pd
from rapidfuzz import fuzz, process

df = pd.read_csv("./edgartools-data.csv")

with open("./predictions.jsonl", "r", encoding="utf-8") as f:
    predictions = [line.strip() for line in f]

with open("./text.jsonl", "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        search = line[57:-365].strip()

        # Find best match
        result = process.extractOne(search, predictions, scorer=fuzz.ratio)
        if result and result[1] > 80:  # score threshold
            match, score, idx = result
            print(f"Line {i} -> prediction {idx} (score: {score})")
            break
```
