```python
import os
from pathlib import Path

import pandas as pd
```

```python
data = []
```

```python
for root, _, files in os.walk("./edgartools-data"):
    for file in files:
        path = os.path.join(root, file)
        path_parts = Path(path).parts
        if path_parts[-2] == "A":
            filing = path_parts[-3] + "/A"
            ticker = path_parts[-4]
        else:
            filing = path_parts[-2]
            ticker = path_parts[-3]
        date = path_parts[-1]
        data.append({"path": path, "ticker": ticker, "filing": filing, "date": date})

df = pd.DataFrame(data)
```

```python
df.to_csv("./edgartools-data.csv", index=False)
```
